#### matplotlib로 그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
# 선그래프 그리기
plt.plot(
    [1,2,3,4,5],
    [1,4,9,15,25]
)

plt.show()
```

```python
# 산점도 그리기
plt.scatter(
    [1,2,3,4,5],
    [1,4,9,15,25]
)

plt.show()
```

```python
x = np.random.randn(1000) # 표준 정규분포(평균:0, 표준편차:1)를 따르는 난수 생성 
y = np.random.randn(1000)

plt.scatter(
    x,
    y
)
plt.show()
```
