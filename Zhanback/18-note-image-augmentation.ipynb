#### Rotation과 Flip을 이용한 이미지 데이터 증강

```python
import os

os.mkdir("./Data")
os.mkdir("./Data/catImage")
```

```python
import numpy as np
```

```python

counter = 0
for angle in range(-15, 15, 1):
    counter+=1
    # 회전 데이터 추가
    img2 = cat.rotate(angle)
    img2.save(f"./Data/catImage/rotation_{counter:02d}.png")

    # Flip 데이터 추가
    img3 = img2.transpose(Image.FLIP_LEFT_RIGHT)
    img3.save(f"./Data/catImage/flip_{counter:02d}.png")
```
