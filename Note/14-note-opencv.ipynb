### 원하는 사이즈의 이미지 생성하기

```python
# opencv는 numpy의 ndarray로 이미지를 표현 
import numpy as np
```

```python
# Color Image 생성
img1 = np.zeros((200, 300, 3), dtype=np.uint8)

# Image 보기
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Color Image 생성
img2 = np.ones((200, 300, 3), dtype=np.uint8) * 255

# Image 보기
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Color Image 녹색 컬러 생성
img2 = np.full((200, 300, 3), (0, 255, 0), dtype=np.uint8)

# Image 보기
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

### 이미지 밝기 조절

```python
# 흑백 이미지의 밝기 조절
src = cv2.imread("/kaggle/input/data-cat/cat.png", cv2.IMREAD_GRAYSCALE)
dst = cv2.add(src, 50)

_, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
axs[1].imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 컬러 이미지의 밝기 조절
src = cv2.imread("/kaggle/input/data-cat/cat.png", cv2.IMREAD_COLOR)
dst = cv2.add(src, (30, 30, 30, 0)) # BGRA

_, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
axs[1].imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
