```python
# 1. 리스트
numbers = [1,2,3,4,5]

# 함수
def square(n):
    return n**2

list(map(square, numbers))
```

```python
# 좀 더 간결하게
numbers = [1,2,3,4,5]

list(map(lambda n: n**2, numbers))
```

```python
# 여러 리스트를 동시에 
list1 = [1,2,3]
list2 = [4,5,6]

list(map(lambda x,y: x+y, list1, list2))
```

```python
# 문자열 리스트 변환 (대문자로 변환)
words = ['hello', 'world', 'python']
list(map(str.upper, words))
```

```python
# 조건을 적용한 변환 (홀수, 짝수)
numbers = [1,2,3,4,5]
list(map(lambda x: "짝수" if x%2 == 0 else "홀수", numbers))
```
