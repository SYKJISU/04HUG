### 1. 기본적인 클래스와 객체 생성

```python
# 클래스 생성 : 생성자를 통한 데이터 전달 
class Person:
    def __init__(self, name, age): # 생성자
        self.name = name
        self.age = age
    def introduce(self):
        return f"안녕하세요 저는 {self.name}이고 {self.age}살 입니다."
```

```python
# 객체생성
p1 = Person("James", 21)
print(p1.introduce())
```

```python
# 클래스 생성 : Property를 통한 데이터 전달 
class Person1:
    # Property, Attribute
    name = None
    age = None
    def __init__(self):
        print("Start")
    def introduce(self):
        return f"안녕하세요 저는 {self.name}이고 {self.age}살 입니다."
```

```python
p2 = Person1()
p2.name = "james"
p2.age = 21
print(p2.introduce())
```

```python
# 클래스 생성 : 함수를 통한 데이터 전달 
class Person2:
    def __init__(self):
        print("Start")
    def introduce(self, name, age):
        return f"안녕하세요 저는 {name}이고 {age}살 입니다."
```

```python
p3 = Person2()
p3.introduce("james", 21)
```

### 상속과 오버라이딩(Overriding)

```python
class Animal:
    def speak(self):
        return "소리를 냅니다"
```

```python
animal = Animal()
animal.speak()
```

```python
class Dog(Animal):
    def speak2(self):
        return "멍멍"
```

```python
dog = Dog()
print(dog.speak())
print(dog.speak2())
```

```python
# Calc라는 클래스
# 기능 : 덧셈, 뺄셈, 곱셈, 나눗셈을 한번에 return 한다.
# Calc생성시 두개의 정수를 입력 받는다

class Calc:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def operand(self):
        return self.num1+self.num2, self.num1-self.num2, self.num1*self.num2, self.num1/self.num2
```

```python
# 생성자를 통한 방법 
calc = Calc(10, 20)
calc.operand()
# (30, -10, 200, 0.5)
```

```python
calc2 = Calc(20, 30)
calc2.operand()
```
