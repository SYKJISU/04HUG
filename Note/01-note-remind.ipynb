```python
# if문 : 짝수 홀수 판별
# 조건 : 사용자로 부터 숫자를 입력받아 숫자가 짝수 인지 홀수 인지 판별하여 출력하기 

num = int(input("숫자를 입력하세요"))

if num % 2 == 0:
    print(f"{num}은 짝수 입니다.")
else:
    print(f"{num}은 홀수 입니다.")
```

```python
# if문 : 학점 변환기 (시험점수 : 0 ~ 100점)
# 조건 : 90점이상 -> A, 80점이상 -> B, 70점이상 -> C, 60점이상-> D, 기타 -> F

score = int(input("점수를 입력하세요 :"))
grade = "" # 학점 

if score > 100 or score < 0:
    print("점수를 확인 하세요")
else:
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"

    print(f"당신의 학점은 {grade}학점 입니다.")
```

```python
# for문 : 1부터 n까지의 합계 구하기 
# 조건 : 사용자가 마지막 숫자를 입력한다. 
# 출력예)
# 마지막 숫자를 입력하세요 : 100
# 1부터 100까지의 합계는 5050 입니다. 

n = int(input("마지막 숫자를 입력 하세요"))
total = 0 # 숫자의 합계

for i in range(1, n+1):
    total += i

print(f"1부터 {n}까지의 합계는 {total} 입니다.")
```

```python
# for문 : 구구단 출력
# 출력예)
# 출력할 구구단의 단을 입력 하세요: 5
dan = int(input("출력할 구구단의 단을 입력 하세요:"))

for i in range(1, 9+1):
    print(f"{dan} X {i} = {dan*i:2d}") # d: decimal, f:float, s:string
```

```python
# collection(List) : 평균 구하기
# 조건 : for문 사용금지, Package(numpy, pandas)사용금지 
# 출력예)
# 점수목록 : [80, 90, 85, 70, 95]
# 평균점수 : 84.00

scores = [80, 90, 85, 70, 95]
total = sum(scores)
average = total / len(scores)

print(f"점수목록 :{scores}")
print(f"평균점수 :{average:.2f}")
```

```python
# collection(Dictionary) : 학생점수 관리
# 출력예1)
# 학생 이름을 입력하세요 : Alice
# Alice의 점수는 85점 입니다.
# 출력예2)
# 학생 이름을 입력하세요 : James
# 해당 학생이 없습니다. 

students = {
    "Alice" : 85,
    "Bob" : 90,
    "Charlie" : 78,
    "David" : 92
}

name = input("학생 이름을 입력 하세요")

if name in students:
    print(f"{name}의 점수는 {students[name]} 입니다.")
else:
    print("해당 학생이 없습니다.")
```

```python
# function : 최대값과 최소값 찾기(for문 사용하기)
def find_min_max(numbers):
    if not numbers: # 리스트가 비어있을 경우
        return None, None
    min_num = numbers[0]
    max_num = numbers[0]

    for number in numbers:
        if number < min_num:
            min_num = number
        if number > max_num:
            max_num = number
            
    return min_num, max_num


nums = [34, 78, 12, 90, 5, 67, 23]
nums1 = []
min_value, max_value = find_min_max(nums1)
print(f"최소값: {min_value}, 최대값: {max_value}")
```

```python
# function : 양수, 음수 갯수 세기
def count_positive_negative(numbers):
    positive_count = negative_count = 0
    for number in numbers:
        if number > 0:
            positive_count += 1
        elif number < 0:
            negative_count += 1
    return positive_count, negative_count

nums = [-10, 15, -20, -30, -5, 40, 0, 25]
pos_count, neg_count = count_positive_negative(nums)
print(f"양의 갯수 :{pos_count}, 음의 갯수 :{neg_count}")
```

```python
# function : 학생 성적을 받아서 합격/불합격(60점 이상) 분류

def classify_students(scores):
    results = []
    no = 0
    for score in scores:
        no += 1
        if score >= 60:
            results.append(f"{no} : 합격")
        else:
            results.append(f"{no} : 불합격")
    return results

student_scores = [85, 40, 73, 90, 58, 62, 33]
print(f"학생 성적 결과 : {classify_students(student_scores)}")
# 학생 성적 결과 : 
# ['1 : 합격', '2 : 불합격', '3 : 합격', '4 : 합격', '5 : 불합격', '6 : 합격', '7 : 불합격']
```

```python
# function : 리스트에서 특정 문자열이 포함된 단어 찾기

def find_words(words, keyword):
    results = []
    for word in words:
        if keyword in word:
            results.append(word)
    return results


word_list = ['apple', 'banana', 'grape', 'pineapple', 'orange']
search = input("검색할 단어를 입력 하세요 :") # app
print(f"{search}가 포함된 단어들 : {find_words(word_list, search)}") # apple, pineapple
```

```python
# function : 단어 갯수 세기 

def word_count_list(text):
    words = text.split()
    unique_words = []
    counts = []
    for word in words:
        if word in unique_words:
            index = unique_words.index(word) # 단어의 index 위치 찾기
            counts[index]+=1
        else:
            unique_words.append(word)
            counts.append(1)
    return list(zip(unique_words, counts))


sentences = input("문장을 입력 하세요 :") # I have have a a dream
print("단어 갯수 :", word_count_list(sentences)) 
# [('I':1), ('have':2), ('a':2), ('dream': 1)]
```
